```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('13-WineQT.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
data.columns
```

```python
plt.scatter("fixed acidity", "quality", data = data)
plt.xlabel("fixed acidity")
plt.ylabel("quality")
plt.title("quality vs fixed acidity")
plt.show()
```

```python
plt.figure(figsize=(26,5))
sns.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Data Value")
#plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(13,8))

plt.subplot(1,2,1)
sns.boxplot(x="total sulfur dioxide", y="quality", data=data)
plt.title("quality by total sulfur dioxide")
plt.xlabel("total sulfur d.")
plt.ylabel("Quality")
```
